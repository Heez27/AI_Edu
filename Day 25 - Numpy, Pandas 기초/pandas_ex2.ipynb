```python
# 데이터 프레임에 한 행을 추가하려고 할때 
# dict 타입으로
df.append({'기관명':'광진구2','소계':420,'2013년도 이전':100,'2014년':100,'2015년':100,'2016년':120}, ignore_index=True)
```

```python
# column을 추가할 때
# 전부 동일한 값이 추가 된다.
df['2021년']=150
```

```python
df
```

```python
# 0과 1행 사이에 행 추가하기
new_data = {'기관명':'구로구2',
            '소계':420,
            '2013년도 이전':100,
            '2014년':100,
            '2015년':100,
            '2016년':120
           }

idx = 1  # 원하는 위치 지정
temp1 = df[df.index <idx]
temp2 = df[df.index >= idx]
df = temp1.append(new_data,ignore_index=True).append(temp2, ignore_index=True)
```
